<h1>Project 2 :Movie Ticket Pricing System 

Description: 
Create a system to calculate movie ticket price based on age and time of show. 
Rules: 
• Age < 5: Free 
• Age 5–12: ₹100 
• Age 13–60: ₹200 
• Age > 60: ₹120 
• If showtime is before 12 PM, apply 20% discount on price

<h4>Movie Ticket Pricing System

```python
# Movie Ticket Pricing System

# Input
age = int(input("Enter your age: "))
show_time = int(input("Enter show time (24-hour format): "))

price = 0

# Age-based pricing
if age < 5:
    price = 0
elif age >= 5 and age <= 12:
    price = 100
elif age >= 13 and age <= 60:
    price = 200
else:
    price = 120

# Apply morning show discount
if show_time < 12:
    price = price * 0.8  # 20% discount

# Output
print("\n--- Ticket Details ---")
print("Age:", age)
print("Show Time:", show_time, ":00")

if price == 0:
    print("Ticket Price: Free")
else:
    print("Ticket Price: ₹", int(price))

```
